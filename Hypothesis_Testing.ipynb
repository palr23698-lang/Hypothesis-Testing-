#Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing ?

Answer - In the context of a t-test, the null hypothesis (H₀) is a formal statement asserting that there is no significant difference between the means of the groups being compared, or that a sample mean is equal to a known population mean. It serves as the baseline assumption against which the alternative hypothesis (H₁) is tested.

For a one-sample t-test, the null hypothesis typically states: H₀: μ = μ₀ where μ is the population mean and μ₀ is the hypothesized mean.

#Question 2: : What does the significance level (α) represent in hypothesis testing?

Answer - In hypothesis testing, the significance level (denoted as α) is a threshold set by the researcher before analyzing data. It represents the maximum probability of making a Type I error — rejecting a true null hypothesis (a false positive) . Common values are 0.05, 0.01, or 0.10, with 0.05 being the most widely used standard.

#Question 3: Differentiate between Type I and Type II errors

Answer - Type I vs Type II Errors
1
In hypothesis testing, two main types of errors can occur when making decisions based on sample data — Type I and Type II errors. Both have distinct meanings, causes, and real-world consequences.

Type I Error (False Positive) happens when a true null hypothesis (H₀) is incorrectly rejected. This means concluding there is an effect or difference when none exists. Example: In medical testing, diagnosing a healthy person as having a disease.

Type II Error (False Negative) occurs when a false null hypothesis is not rejected. This means failing to detect an actual effect or difference. Example: In medical testing, failing to diagnose a patient who actually has the disease.

#Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each ?

Answer - #Difference Between One-Tailed and Two-Tailed Tests

# One-Tailed Test
A one-tailed test is based on a uni-directional hypothesis where the area of rejection is on only one side of the sampling distribution. It determines whether a particular population parameter is larger or smaller than the predefined parameter. It uses one single critical value to test the data.

Example: Effect of participants of students in coding competition on their fear level.
H0: There is no important effect of students in coding competition on their fear level.

#Two-Tailed Test
A two-tailed test is also called a nondirectional hypothesis. For checking whether the sample is greater or less than a range of values, we use the two-tailed. It is used for null hypothesis testing.

Example: Effect of new bill pass on the loan of farmers.
H0: There is no significant effect of the new bill passed on loans of farmers.

#Question 6:When should you use a Z-test instead of a t-test?

Answer - We can use a Z-test instead of a t-test when the population standard deviation is known or when the sample size is large enough that the sample standard deviation closely estimates the population standard deviation.
